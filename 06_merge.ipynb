```python
#Load the departments_t CSV file into a DataFrame 
df_departments_t = pd.read_csv('../data/processed/departments_t.csv')
```

```python
#Merge products with departments_t on 'department_id' using a left join
df_products_departments = pd.merge(
    df_products,
    df_departments_t,
    on = "department_id",
    how = "left",
    indicator = True
)
df_products_departments.head()
```

```python
#Looking at the indicator results
df_products_departments["_merge"].value_counts()
```

```python
#Investigating unmatched rows
df_products_departments[
    df_products_departments["_merge"] == "left_only"
].head()
```

```python
#cleaning the indicator results or rerun the merge without indicator=True
df_products_departments = pd.merge(
    df_products,
    df_departments_t,
    on = "department_id",
    how = "left"
)
df_products_departments.head()
```

```python
# Filter orders to keep only the train dataset
df_oreders_train = df_orders [df_orders["eval_set"] == "train"]
```

```python
#Performing the merge
#We merge the filtered order table with the transaction bridge table using order_id.
df_train_transaction = pd.merge(
    df_oreders_train,
    df_order_products_train,
    on = "order_id",
    how = "left",
)

df_train_transaction.head()
```

```python
#Checking the merge with indicator= True
df_train_transaction_check = pd.merge(
    df_oreders_train,
    df_order_products_train,
    on = "order_id",
    how = "left",
    indicator = True
)
df_train_transaction_check.head()
```

```python
#Inspecting the _merge column
df_train_transaction_check["_merge"].value_counts()
```

```python
#Looking at unmuched orders
df_train_transaction_check[
    df_train_transaction_check["_merge"] == "left_only"
].head()
```

```python
#rerun without indicator= True
df_train_transaction_check = pd.merge(
    df_oreders_train,
    df_order_products_train,
    on = "order_id",
    how = "left"

)
df_train_transaction_check.head()
```

```python
#Checking the row counts
print(df_oreders_train.shape)
print(df_order_products_train.shape)
print(df_train_transaction.shape)
```

```python
#check all columns we have now
df_train_transaction.columns
```

```python
#The last step is to merge df_train_transactions with df_products_enriched.
df_train_analytic = pd.merge(
    df_train_transaction,
    df_products_enriched,
    on = "product_id",
    how = "left"
)
df_train_analytic.head()
```

```python
#Checking the final merge with indicator=True
df_train_analytic = pd.merge(
    df_train_analytic,
    df_products_enriched,
    on = "product_id",
    how = "left",
    indicator = True
)
df_train_analytic.head()
```

```python
#Inspecting the _merge column
df_train_analytic["_merge"].value_counts()
```

```python
#Investigating unmatched product rows
df_train_analytic[
    df_train_analytic["_merge"] == "left_only"
].head()
```

```python
#rerun without indicator=True
df_train_analytic = pd.merge(
    df_train_analytic,
    df_products_enriched,
    on = "product_id",
    how = "left"
)
df_train_analytic.head()
```
