#### Homework

```python
#Given
revenues = [120, 250, 80, 310, 95]
tax_rate = 0.18
discount_rate = 0.10
```

```python
#1.Compute revenues including tax.
revenues_with_tax = [r * (1 + tax_rate) for r in revenues]

print(revenues_with_tax)
```

```python
#2.Compute final revenues after tax and discount.
final_revenues = [r * (1 + tax_rate) * (1 - discount_rate) for r in revenues]

print(final_revenues)
```

```python
#3.Create a Boolean list indicating whether revenue > 100.
greater_than_100 = [r > 100 for r in revenues]

print(greater_than_100)
```

```python
#4.Create a Boolean list indicating whether revenue is between 100 and 300.
between_100_300 = [100 < r < 300 for r in revenues]

print(between_100_300)
```

#### 5.Why do parentheses matter in the formula?
Parentheses control the order of operations in mathematical expressions.

In the revenue calculation we use:

Final Revenue = R * (1 + tax_rate) * (1 - discount_rate)

The parentheses ensure that the tax and discount adjustments are calculated first before multiplying the revenue.

Without parentheses, Python could apply operations in a different order, which would produce incorrect financial results.

#### Part 3.
- Segmentation rules:
- “Premium” if revenue > 250
- “Standard” if 100 < revenue ≤ 250
- “Low” otherwise

```python
#### 1.Create a list of segment labels.
segments = []

for s in sales:
    if s > 250:
        segments.append("Premium")
    elif 100 < s <= 250:
        segments.append("Standard")
    else:
        segments.append("Low")

print(segments)
```

```python
# 2.Count how many customers fall into each segment.
premium_count = 0
standard_count = 0
low_count = 0
```

```python
for seg in segments:
    if seg == "Premium":
        premium_count += 1
    elif seg == "Standard":
        standard_count += 1
    else:
        low_count += 1

print("Premium:", premium_count)
print("Standard:", standard_count)
print("Low:", low_count)
```

##### 3.Why does the order of elif statements matter?
The order of elif statements matters because Python evaluates conditions from top to bottom.

Once a condition is True, Python executes that block and skips the remaining conditions.

For example, if we check revenue > 100 before revenue > 250,
values greater than 250 would already match the first condition
and would never reach the Premium category.

##### 4.Why sets are unordered.
Sets in Python are unordered collections because they are implemented using hash tables.

Elements in a set are stored based on their hash values rather than their insertion order. 
Because of this internal structure, Python does not guarantee any specific order when elements are stored or displayed.

This design allows sets to efficiently perform operations like removing duplicates and checking membership.

##### 5.Why are dictionaries useful for structured data?

Dictionaries are useful for structured data because they store information as key-value pairs.

Dictionaries are widely used in data analytics because they allow fast lookups and flexible data organization.

##### 4.Compare list comprehension vs loop in Markdown.
List comprehension is a concise way to create lists in Python. 
It allows us to perform operations, apply conditions, and generate new lists in a single readable line.

Using loops requires multiple lines of code where we initialize a list, iterate through elements, and append results manually.

List comprehension is often preferred in data analytics because it is more compact and usually easier to read when performing simple transformations or filtering operations.

However, traditional loops can be more flexible and easier to understand for complex logic.

#### Part 7 Convert our data into a DataFrame.

```python
import pandas as pd

data = {
    "revenue": revenues
}

df = pd.DataFrame(data)
df
```

```python
#1.Add column "revenue_after_tax".
df["revenue_after_tax"] = df["revenue"] * 0.9

df
```

```python
#2.Add column "segment" using conditional logic.
def segment_label(r):
    if r > 250:
        return "Premium"
    elif 100 < r <= 250:
        return "Standard"
    else:
        return "Low"

df["segment"] = df["revenue"].apply(segment_label)

df
```

```python
#3.Filter rows where revenue > 100.
filtered_df = df[df["revenue"] > 100]

filtered_df
```

```python
#4.Remove the "revenue_after_tax" column.
df = df.drop(columns=["revenue_after_tax"])

df
```

##### 6.Why is this easier than manual loops?
Using pandas DataFrames makes data analysis easier because operations can be applied to entire columns at once.

Instead of writing manual loops to process each value individually, pandas allows vectorized operations that are faster and require less code.

DataFrames also provide built-in tools for filtering, aggregation, and transformation, making data manipulation more efficient and easier to read.

### Bonus Reflection

**1. Difference between mutable and immutable objects**  
Mutable objects (like lists, dictionaries) can be changed in place, while immutable objects (like tuples, strings) cannot be modified after creation.

**2. Why loops are less efficient than pandas vectorization**  
Loops process data element by element, which is slow in Python. Pandas vectorization applies operations over entire columns at once, leveraging optimized C-level code for speed.

**3. How Boolean logic relates to SQL WHERE**  
Boolean logic (AND, OR, NOT) determines which rows meet conditions. In SQL, the `WHERE` clause filters rows using these logical expressions.

**4. Why understanding lists helps understand DataFrames**  
DataFrames are like tables of lists (columns). Knowing how lists store and manipulate data makes it easier to work with columns, apply operations, and understand indexing in DataFrames.
