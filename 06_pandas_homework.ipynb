```python
# Import pandas library for data manipulation
import pandas as pd
```

```python
# Define URLs of the datasets that need to be downloaded
URLS = [
    "https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/raw/customers.csv",
    "https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/raw/states.csv"
]
```

```python
# Define filenames for saving the datasets locally
names = ["customers", "states"]
```

```python
# Download customers dataset from URL
df_customers = pd.read_csv(
"https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/raw/customers.csv"
)

# Save customers dataset into raw folder
df_customers.to_csv("../data/raw/customers.csv", index=False)
```

```python
# Download states dataset from URL
df_states = pd.read_csv(
"https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/raw/states.csv"
)

# Save states dataset into raw folder
df_states.to_csv("../data/raw/states.csv", index=False)
```

```python
# Load datasets from raw folder

df_orders = pd.read_csv("../data/raw/orders.csv")
df_customers = pd.read_csv("../data/raw/customers.csv")
df_states = pd.read_csv("../data/raw/states.csv")
```

```python
# Check first rows of orders dataset
df_orders.head()
```

```python
df_orders.columns
```

```python
# Merge orders dataset with customers dataset using user_id
df = df_orders.merge(df_customers, on="user_id", how="left")
```

```python
# Check first rows of merged dataset
df.head()
```

```python
# Check dataset shape
df.shape
```

```python
# Rename 'STATE' column to 'state' to match states dataset
df = df.rename(columns={"STATE": "state"})
```

```python
# Merge the dataframe with states dataset on 'state'
df = df.merge(df_states, on="state", how="left")
```

```python
# Check first rows
df.head()
```

```python
# Check dataframe shape
df.shape
```

```python
#Save final merged dataset into processed folder
df.to_csv("../data/processed/orders_customers_states.csv", index=False)
```
