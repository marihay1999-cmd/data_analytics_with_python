```python
# Import pandas library for data manipulation
import pandas as pd
```

```python
# Define URLs of the datasets that need to be downloaded
URLS = [
    "https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/raw/customers.csv",
    "https://raw.githubusercontent.com/hovhannisyan91/data_analytics_with_python/refs/heads/main/data/raw/states.csv"
]
```

```python
# Define filenames for saving the datasets locally
names = ["customers", "states"]
```

```python

#Loop through URLs and filenames to download and save each file
for url, name in zip(URLS, names):
    df = pd.read_csv(url)                    # Download CSV from URL
    df.to_csv(f'../data/raw/{name}.csv', index=False)  # Save CSV in raw folder

print("Datasets downloaded and saved in raw folder successfully.")
```

```python
# Load datasets
df_orders = pd.read_csv("../data/raw/orders.csv")            
df_customers = pd.read_csv("../data/raw/customers.csv")       
df_states = pd.read_csv("../data/raw/states.csv")  
```

```python
# Merge orders with customers on 'user_id'
df = df_orders.merge(df_customers, on="user_id", how="left")
```

```python
# Rename 'STATE' to 'state' for consistent merge with states dataset
df = df.rename(columns={"STATE": "state"})
```

```python
# Merge with states on 'state'
df = df.merge(df_states, on="state", how="left")
```

```python
# Optional: check the first rows to make sure merge worked
print(df.head())
```

```python
processed_folder = "../data/processed"  # Path to processed folder

# Save the merged dataset
df.to_csv(f"{processed_folder}/orders_customers_states.csv", index=False)

print("Final merged dataset saved successfully in processed folder.")
```
